```python
# import tensorflow as tf
# import numpy as np
# from sklearn.neighbors import KNeighborsClassifier
# from sklearn.preprocessing import LabelEncoder
# from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# from tensorflow.keras.preprocessing import image
# import matplotlib.pyplot as plt

# # Step 1: Load the dataset
# IMAGE_SIZE = 224  # Adjust based on your image dimensions
# BATCH_SIZE = 32

# # Load dataset from the "PlantVillage" directory
# dataset = tf.keras.preprocessing.image_dataset_from_directory(
#     "PlantVillage",  # Path to your dataset directory
#     seed=123,
#     shuffle=True,
#     image_size=(IMAGE_SIZE, IMAGE_SIZE),
#     batch_size=BATCH_SIZE
# )

# # Get class names (Healthy, Early Blight, Late Blight)
# class_names = dataset.class_names
# print(f"Class Names: {class_names}")

# # Step 2: Split the dataset into training, validation, and test sets
# def get_dataset_partitions_tf(ds, train_split=0.7, val_split=0.1, test_split=0.2, shuffle=True, shuffle_size=10000):
#     assert abs(train_split + val_split + test_split - 1) < 1e-6, "The splits must sum up to 1"
    
#     ds_size = len(ds)
#     if shuffle:
#         ds = ds.shuffle(shuffle_size, seed=12)
    
#     train_size = int(train_split * ds_size)
#     val_size = int(val_split * ds_size)
    
#     train_ds = ds.take(train_size)
#     val_ds = ds.skip(train_size).take(val_size)
#     test_ds = ds.skip(train_size + val_size)
    
#     return train_ds, val_ds, test_ds

# train_ds, val_ds, test_ds = get_dataset_partitions_tf(dataset)

# # Convert TensorFlow dataset to NumPy arrays
# def dataset_to_numpy(ds):
#     images = []
#     labels = []
#     for image_batch, label_batch in ds:
#         images.append(image_batch.numpy())
#         labels.append(label_batch.numpy())
#     return np.concatenate(images), np.concatenate(labels)

# # Convert datasets to NumPy arrays
# train_images, train_labels = dataset_to_numpy(train_ds)
# val_images, val_labels = dataset_to_numpy(val_ds)
# test_images, test_labels = dataset_to_numpy(test_ds)

# # Normalize the images
# train_images = train_images / 255.0
# val_images = val_images / 255.0
# test_images = test_images / 255.0

# # Flatten the images for KNN
# train_images_flat = train_images.reshape(train_images.shape[0], -1)
# val_images_flat = val_images.reshape(val_images.shape[0], -1)
# test_images_flat = test_images.reshape(test_images.shape[0], -1)

# # Encode the labels
# all_labels = np.concatenate((train_labels, val_labels, test_labels))
# encoder = LabelEncoder()
# encoder.fit(all_labels)

# train_labels_encoded = encoder.transform(train_labels)
# val_labels_encoded = encoder.transform(val_labels)
# test_labels_encoded = encoder.transform(test_labels)

# # Combine training and validation sets for KNN training
# X_train = np.concatenate((train_images_flat, val_images_flat), axis=0)
# y_train = np.concatenate((train_labels_encoded, val_labels_encoded), axis=0)

# # Test set
# X_test = test_images_flat
# y_test = test_labels_encoded

# # Step 3: Train KNN Classifier
# knn = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
# knn.fit(X_train, y_train)

# # Step 4: Evaluate the model
# y_pred = knn.predict(X_test)
# accuracy = accuracy_score(y_test, y_pred)
# print(f"KNN Accuracy on Test Set: {accuracy * 100:.2f}%")

# print("\nClassification Report:")
# print(classification_report(y_test, y_pred, target_names=class_names))

# print("Confusion Matrix:")
# print(confusion_matrix(y_test, y_pred))

# # Step 5: Function to preprocess an input image
# def preprocess_image(img_path, target_size=(IMAGE_SIZE, IMAGE_SIZE)):
#     img = image.load_img(img_path, target_size=target_size)
#     img_array = image.img_to_array(img)
#     img_array = img_array / 255.0  # Normalize the image
#     img_flat = img_array.reshape(1, -1)  # Flatten for KNN
#     return img_flat

# # Step 6: Function to predict disease from an external image input
# def predict_disease(img_path):
#     # Preprocess the input image
#     img_flat = preprocess_image(img_path)
    
#     # Make a prediction using the trained KNN model
#     pred_label_encoded = knn.predict(img_flat)
    
#     # Decode the predicted label to the actual class name
#     pred_label = encoder.inverse_transform(pred_label_encoded)[0]  # Get the first element since it's a single prediction
    
#     # Print the predicted disease
#     print(f"Predicted disease: {pred_label}")
    
#     # Show the image with the predicted label as the title
#     img = image.load_img(img_path, target_size=(IMAGE_SIZE, IMAGE_SIZE))
#     plt.imshow(img)
#     plt.title(f"Predicted: {pred_label}")
#     plt.axis("off")
#     plt.show()


# # Step 7: Example usage - Provide the path of any image outside the dataset
# # You can test with any new image from outside the dataset
# image_path = "eb.png"  # Path to your image in the main folder
# predict_disease(image_path)

# import tensorflow as tf
# import numpy as np
# from sklearn.neighbors import KNeighborsClassifier
# from sklearn.preprocessing import LabelEncoder
# from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# from tensorflow.keras.preprocessing import image
# import matplotlib.pyplot as plt

# # Step 1: Load the dataset
# IMAGE_SIZE = 224  # Image size compatible with MobileNetV2
# BATCH_SIZE = 32

# # Load dataset from the "PlantVillage" directory
# dataset = tf.keras.preprocessing.image_dataset_from_directory(
#     "PlantVillage",  # Path to your dataset directory
#     seed=123,
#     shuffle=True,
#     image_size=(IMAGE_SIZE, IMAGE_SIZE),
#     batch_size=BATCH_SIZE
# )

# class_names = dataset.class_names
# print(f"Class Names: {class_names}")

# # Step 2: Split the dataset into training, validation, and test sets
# def get_dataset_partitions_tf(ds, train_split=0.7, val_split=0.1, test_split=0.2):
#     assert abs(train_split + val_split + test_split - 1) < 1e-6, "The splits must sum up to 1"
    
#     ds_size = len(ds)
#     train_size = int(train_split * ds_size)
#     val_size = int(val_split * ds_size)
    
#     train_ds = ds.take(train_size)
#     val_ds = ds.skip(train_size).take(val_size)
#     test_ds = ds.skip(train_size + val_size)
    
#     return train_ds, val_ds, test_ds

# train_ds, val_ds, test_ds = get_dataset_partitions_tf(dataset)

# # Step 3: Load the MobileNetV2 model for feature extraction
# base_model = tf.keras.applications.MobileNetV2(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3),
#                                                include_top=False,
#                                                weights='imagenet')
# base_model.trainable = False  # Freeze the base model

# # Feature extraction function
# def extract_features(dataset):
#     features = []
#     labels = []
#     for batch_images, batch_labels in dataset:
#         batch_features = base_model(batch_images)
#         batch_features = tf.keras.layers.GlobalAveragePooling2D()(batch_features)
#         features.append(batch_features.numpy())
#         labels.append(batch_labels.numpy())
#     return np.concatenate(features), np.concatenate(labels)

# # Step 4: Extract features and encode labels
# train_features, train_labels = extract_features(train_ds)
# val_features, val_labels = extract_features(val_ds)
# test_features, test_labels = extract_features(test_ds)

# # Encode the labels
# encoder = LabelEncoder()
# encoder.fit(np.concatenate((train_labels, val_labels, test_labels)))

# train_labels_encoded = encoder.transform(train_labels)
# val_labels_encoded = encoder.transform(val_labels)
# test_labels_encoded = encoder.transform(test_labels)

# # Combine training and validation sets for KNN training
# X_train = np.concatenate((train_features, val_features), axis=0)
# y_train = np.concatenate((train_labels_encoded, val_labels_encoded), axis=0)

# # Step 5: Train the KNN Classifier
# knn = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
# knn.fit(X_train, y_train)

# # Step 6: Evaluate the model
# y_pred = knn.predict(test_features)
# accuracy = accuracy_score(test_labels_encoded, y_pred)
# print(f"KNN Accuracy on Test Set: {accuracy * 100:.2f}%")

# print("\nClassification Report:")
# print(classification_report(test_labels_encoded, y_pred, target_names=class_names))

# print("Confusion Matrix:")
# print(confusion_matrix(test_labels_encoded, y_pred))

# # Step 7: Function to preprocess an input image for feature extraction
# def preprocess_image(img_path):
#     img = image.load_img(img_path, target_size=(IMAGE_SIZE, IMAGE_SIZE))
#     img_array = image.img_to_array(img)
#     img_array = tf.expand_dims(img_array, axis=0)
#     img_array = tf.keras.applications.mobilenet_v2.preprocess_input(img_array)
#     img_features = base_model(img_array)
#     img_flat = tf.keras.layers.GlobalAveragePooling2D()(img_features).numpy()
#     return img_flat

# # Step 8: Function to predict disease from an external image input
# def predict_disease(img_path):
#     # Preprocess the input image
#     img_features = preprocess_image(img_path)
    
#     # Make a prediction using the trained KNN model
#     pred_label_encoded = knn.predict(img_features)
    
#     # Decode the predicted label to the actual class name
#     pred_label = encoder.inverse_transform(pred_label_encoded)[0]
    
#     # Print the predicted disease
#     print(f"Predicted disease: {pred_label}")
    
#     # Show the image with the predicted label as the title
#     img = image.load_img(img_path, target_size=(IMAGE_SIZE, IMAGE_SIZE))
#     plt.imshow(img)
#     plt.title(f"Predicted: {pred_label}")
#     plt.axis("off")
#     plt.show()

# # Step 9: Example usage - Provide the path of any image outside the dataset
# image_path = "plbb.jpeg"  # Path to your external image
# predict_disease(image_path)

import tensorflow as tf
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns

# Step 1: Load the dataset
IMAGE_SIZE = 224  # Image size compatible with MobileNetV2
BATCH_SIZE = 32

dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "PlantVillage",
    seed=123,
    shuffle=True,
    image_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE
)

class_names = dataset.class_names
print(f"Class Names: {class_names}")

# Step 2: Split the dataset into training, validation, and test sets
def get_dataset_partitions_tf(ds, train_split=0.7, val_split=0.1, test_split=0.2):
    assert abs(train_split + val_split + test_split - 1) < 1e-6, "Splits must sum to 1"
    
    ds_size = len(ds)
    train_size = int(train_split * ds_size)
    val_size = int(val_split * ds_size)
    
    train_ds = ds.take(train_size)
    val_ds = ds.skip(train_size).take(val_size)
    test_ds = ds.skip(train_size + val_size)
    
    return train_ds, val_ds, test_ds

train_ds, val_ds, test_ds = get_dataset_partitions_tf(dataset)

# Step 3: Load MobileNetV2 for feature extraction
base_model = tf.keras.applications.MobileNetV2(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3),
                                               include_top=False,
                                               weights='imagenet')
base_model.trainable = False  # Freeze the base model

# Feature extraction function
def extract_features(dataset):
    features, labels = [], []
    for batch_images, batch_labels in dataset:
        batch_features = base_model(batch_images)
        batch_features = tf.keras.layers.GlobalAveragePooling2D()(batch_features)
        features.append(batch_features.numpy())
        labels.append(batch_labels.numpy())
    return np.concatenate(features), np.concatenate(labels)

# Extract features and encode labels
train_features, train_labels = extract_features(train_ds)
val_features, val_labels = extract_features(val_ds)
test_features, test_labels = extract_features(test_ds)

# Encode the labels
encoder = LabelEncoder()
encoder.fit(np.concatenate((train_labels, val_labels, test_labels)))

train_labels_encoded = encoder.transform(train_labels)
val_labels_encoded = encoder.transform(val_labels)
test_labels_encoded = encoder.transform(test_labels)

# Combine training and validation sets for Logistic Regression training
X_train = np.concatenate((train_features, val_features), axis=0)
y_train = np.concatenate((train_labels_encoded, val_labels_encoded), axis=0)

# Step 4: Visualize features with PCA
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)

plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_train_pca[:, 0], y=X_train_pca[:, 1], hue=y_train, palette="viridis", style=y_train, markers=["o", "s", "D"])
plt.title("PCA of Extracted Features")
# plt.show()

# Step 5: Train Logistic Regression Classifier
logreg = LogisticRegression(max_iter=1000, multi_class='auto', solver='lbfgs', n_jobs=-1)
logreg.fit(X_train, y_train)

# Step 6: Evaluate the model
y_pred = logreg.predict(test_features)
accuracy = accuracy_score(test_labels_encoded, y_pred)
print(f"Logistic Regression Accuracy on Test Set: {accuracy * 100:.2f}%")

print("\nClassification Report:")
print(classification_report(test_labels_encoded, y_pred, target_names=class_names))

print("Confusion Matrix:")
print(confusion_matrix(test_labels_encoded, y_pred))

# Step 7: Function to preprocess and predict disease from an external image input
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(IMAGE_SIZE, IMAGE_SIZE))
    img_array = image.img_to_array(img)
    img_array = tf.expand_dims(img_array, axis=0)
    img_array = tf.keras.applications.mobilenet_v2.preprocess_input(img_array)
    img_features = base_model(img_array)
    img_flat = tf.keras.layers.GlobalAveragePooling2D()(img_features).numpy()
    return img_flat

def predict_disease(img_path):
    img_features = preprocess_image(img_path)
    pred_label_encoded = logreg.predict(img_features)
    pred_label = encoder.inverse_transform(pred_label_encoded)[0]
    
    print(f"Predicted disease: {pred_label}")
    img = image.load_img(img_path, target_size=(IMAGE_SIZE, IMAGE_SIZE))
    plt.imshow(img)
    plt.title(f"Predicted: {pred_label}")
    plt.axis("off")
    plt.show()

# Example usage
image_path = "eb.png"  # Path to external image
predict_disease(image_path)

 

```
